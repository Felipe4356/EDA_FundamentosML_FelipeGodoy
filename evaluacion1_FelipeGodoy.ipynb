```python
# Configuración del gráfico
plt.figure(figsize=(12, 6))
sns.boxplot(x=df['MntFruits'])
plt.xticks(rotation=45)
plt.title("Boxplot de estadísticas ")
plt.show()

# Calcular cuartiles
Q1 = df['MntFruits'].quantile(0.25)
Q2 = df['MntFruits'].median()
Q3 = df['MntFruits'].quantile(0.75)
IQR = Q3 - Q1
limite_superior = Q3 + 1.5 * IQR

print(f"Q1: {Q1}, Q2 (mediana): {Q2}, Q3: {Q3}")
```
