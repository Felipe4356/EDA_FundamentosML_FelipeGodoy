# **6.Observacion de probabilidades**
aqui haremos  dos probalidad con dos preguntas:

1.Â¿cual seria la probalidad de que un cliente tenga un gasto en frutas superior a 100 unidades?

2.Â¿cual seria la probabilidad que un cliente sea soltero y tenga un gastos  en frutas superior a 50 unidades?

```python

clientes_gasto_alto = len(df[df['MntFruits'] > 100])

# Calcular la cantidad total de clientes.
total_clientes = len(df)

# Calcular la probabilidad.
probabilidad = clientes_gasto_alto / total_clientes

print(f"Probabilidad : {probabilidad:.4f}")
```

```python

# Filtrar los clientes solteros.
solteros = df[df['Marital_Status'] == 'Single']

# Filtrar los clientes solteros con un gasto en frutas superior a 50.
solteros_gasto_alto = solteros[solteros['MntFruits'] > 50]

# Calcular la cantidad de clientes solteros con gasto alto en frutas.
cantidad_solteros_gasto_alto = len(solteros_gasto_alto)

# Calcular la cantidad total de clientes solteros.
cantidad_total_solteros = len(solteros)

# Calcular la probabilidad.
probabilidad = cantidad_solteros_gasto_alto / cantidad_total_solteros

print(f"Probabilidad : {probabilidad:.4f}")
```

resultado:

P1: La probabilidad que un cliente tenga un gasto en frutas superior a 100 unidades son de 0.0759

P2: La probabilidad de que un cliente sea soltero y tenga un gasto en frutas superior a 50 unidades es: 0.1699

# **7.Grafico de Regresion lineal y matriz de correlacion**

```python

df_cleaned = df.dropna(subset=['Income'])
plt.figure(figsize=(10, 6))
sns.regplot(x='Income', y='MntFruits', data=df_cleaned)

plt.xlabel('Income')
plt.ylabel('MntFruits')
plt.title('Regresion lineal: ingresos y gastos en frutas por unidad')
plt.show()
```

Resultado:

en el grafico tiene una tendencia positiva en los ingresos y gastos en frutas igual se puede ver que hay valores atipicos  presentes  en el grafico
